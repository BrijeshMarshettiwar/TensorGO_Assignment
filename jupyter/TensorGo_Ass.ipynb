```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_data2.csv')
```

```python
print(df)
```

```python
print(df['fees'].mean())
```

```python
print(df['fine'].mean())
```

```python
print(df['fees'].median())
```

```python
print(df['fine'].median())
```

```python
print(df['fine'].mode())
```

```python
print(df['fine'].std())
```

```python
print(df['fine'].corr(df['fees']))
```

```python
df['fees'].hist(bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Fees')
plt.ylabel('Frequency')
plt.title('Histogram of Fees')
plt.show()
```

```python
plt.scatter(df['fees'], df['fine'], color='blue', edgecolor='black')
plt.xlabel('Fees')
plt.ylabel('Fine')
plt.title('Scatter Plot of Fees vs Fine')
plt.show()
```

```python
plt.plot(df['name'], df['fees'], marker='o', color='blue', linestyle='-', label='Fees')
plt.plot(df['name'], df['fine'], marker='o', color='green', linestyle='--', label='Fine')
plt.xlabel('Name')
plt.ylabel('Amount')
plt.title('Line Plot of Fees and Fine')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
